```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('../input/foodpricespakistandata/wfp_food_prices_pakistan_v2.csv', sep=';')
data.info()
```

```python
# some first lines
data.head(10)
```

```python
data['date'] =  pd.to_datetime(data['date'], format='%d/%m/%Y')
```

```python
data.head(10)
```

```python
# unique values for some object columns
unique_obj_values = {}
obj_cols = list(data.select_dtypes('object').columns)
for c in obj_cols:
    unique_obj_values[c] = list(data[c].unique())
    
print(unique_obj_values)
```

```python
unique_obj_values['cmname']
```

# Let's look at the structure of data related to Wheat flour - Retail

```python
Wheat_flour_Retail = data.loc[data['cmname'] == 'Wheat flour - Retail', :]

Wheat_flour_Retail.info()
```

```python
Wheat_flour_Retail
```

```python
print('Start Date : ', Wheat_flour_Retail.date.min())
print('End Date : ', Wheat_flour_Retail.date.max())
print('Number of unique Dates : ', len(Wheat_flour_Retail.date.unique()))
```

```python
wfr_prices_by_market = Wheat_flour_Retail[['date', 'price', 'mktname']].pivot(index='date', columns='mktname', values='price')
wfr_prices_by_market = pd.DataFrame(data = wfr_prices_by_market.reset_index().values, columns = ['date', 'Karachi', 'Lahore', 'Multan', 'Peshawar','Quetta'])
wfr_prices_by_market = wfr_prices_by_market.set_index('date')
wfr_prices_by_market
```

```python
# Apply the default theme
fig, ax = plt.subplots(figsize=(20, 5))

sns.lineplot(data = wfr_prices_by_market,  ax=ax)
plt.show()
```
